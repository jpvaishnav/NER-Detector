```python
#Stanford NER begnis here
import nltk
nltk.download('punkt')
from nltk.tag.stanford import StanfordNERTagger

"""
Named Entity tagging in Python with NLTK and the Stanford NER tagger
"""
f=open(r"/content/drive/MyDrive/Colab Notebooks/NLP_2/data.txt","r")
content=f.read()


PATH_TO_JAR='/content/drive/MyDrive/Colab Notebooks/NLP_2/stanford-ner.jar'
PATH_TO_MODEL = '/content/drive/MyDrive/Colab Notebooks/NLP_2/english.all.3class.distsim.crf.ser.gz'


tagger = StanfordNERTagger(model_filename=PATH_TO_MODEL,path_to_jar=PATH_TO_JAR, encoding='utf-8')

sentence = content

#split the sentence into words
words = nltk.word_tokenize(sentence) 

tagged = tagger.tag(words)
# print(tagged)

for result in tagged:
    tag_value = result[0]
    tag_type = result[1]
    if tag_type != 'O':
      print('Type: %s, Value: %s' % (tag_type, tag_value))
```
